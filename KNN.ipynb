Classificação Baseada em Vizinhança

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
carros=pd.read_csv('carros_modelos_categorical.csv')
display(carros)
```

```python
carros['label']=np.where(carros['popularidade']>carros['popularidade'].mean(),1,0)
y=carros['label'].to_numpy()
carros=carros.drop(columns=['fabricante','modelo','popularidade','ano'])
X=carros.iloc[:, :-1]
y=carros.iloc[:,-1]
X=StandardScaler().fit_transform(X)
```

```python
X_treino,X_teste,y_treino,y_teste=train_test_split(X,y,test_size=0.20,stratify=y,
                                                   random_state=123,shuffle=True)
X_treino,X_teste,y_treino,y_teste=train_test_split(X,y,random_state=0)
knn5=KNeighborsClassifier(algorithm='ball_tree') #k=5
knn1=KNeighborsClassifier(n_neighbors=1) #k=1
knn3=KNeighborsClassifier(n_neighbors=3) #k=3
knn5.fit(X_treino,y_treino)
knn1.fit(X_treino,y_treino)
knn3.fit(X_treino,y_treino)
y_pred5=knn5.predict(X_teste)
y_pred1=knn1.predict(X_teste)
y_pred3=knn3.predict(X_teste)

```

```python
from sklearn.metrics import accuracy_score
print('Acurácia com k=5',accuracy_score(y_teste,y_pred5))
print('Acurácia com k=3',accuracy_score(y_teste,y_pred3))
print('Acurácia com k=1',accuracy_score(y_teste,y_pred1))
```

```python
from sklearn.metrics import classification_report
categorias=['Popular','Não-Popular']
print(classification_report(y_teste,y_pred5,target_names=categorias))
```

```python
print(classification_report(y_teste,y_pred3,target_names=categorias))
```

```python
from sklearn.model_selection import GridSearchCV
knn_grid=GridSearchCV(estimator=KNeighborsClassifier(),
                      param_grid={'n_neighbors': np.arange(1,20)},cv=5)
knn_grid.fit(X_treino,y_treino)
y_pred=knn_grid.predict(X_teste)
from sklearn.metrics import accuracy_score
print('Acurácia com k=5',accuracy_score(y_teste,y_pred)) 
```

```python
#Observando os outros parâmetros
print(classification_report(y_teste,y_pred,target_names=categorias))
```
